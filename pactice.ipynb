```python
import pandas as pd 

data = {
    "Name": ["Aban", "Apil", "Alit", "Shan", "Om", "Riya", "Khan", "Mitu"],
    "Age": [25, None, 44, 23, None, 29, 35, None],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR"],
    "Salary": [50000, 60000, 70000, None, None, 65000, 72000, 58000],
    "Experience": [2, 4, 10, 1, None, 5, 8, 3]
}

df = pd.DataFrame(data)

print(df)

#check empty by number
print(df.isnull().sum())

#check empty by percentage 
print(df.isnull().mean()* 100)

#ignore unfinshed data
print(df.dropna())

#Nan is fill with mean value
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary']).mean()
df['Experience'] = df['Experience'].fillna(df['Experience'].mean())

print(df)
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

data = {
    "Student_ID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Ayan","Ritu","Kabir","Nila","Sohan","Maya","Arif","Tina","Rahim","Puja"],
    "Gender": ["Male","Female","Male","Female","Male","Female","Male","Female","Male","Female"],
    "Course": ["Python","Data Science","Python","ML","ML","Python","Data Science","Python","ML","Data Science"],
    "Level": ["Beginner","Intermediate","Beginner","Advanced","Intermediate","Beginner","Advanced","Intermediate","Advanced","Beginner"],
    "Hours_Studied": [20,35,15,50,40,None,60,30,None,25],
    "Score": [65,78,55,88,82,70,90,None,85,60],
    "Completed": ["Yes","Yes","No","Yes","Yes","No","Yes","Yes","No","Yes"]
}

df = pd.DataFrame(data)


#lebel_encoding 

le= LabelEncoder()
df_level = df.copy()

df_level['Gender_encoded'] = le.fit_transform(df_level['Gender'])
df_level['Completed_encoded']= le.fit_transform(df_level['Completed'])

print('\n Lavel Encoding')
print(df_level[['Name','Gender','Gender_encoded', 'Completed','Completed_encoded']])


# One_Hot Encoding 
 
df_encoding = pd.get_dummies(df_level, columns=['Course', 'Level'], dtype= int )
print(df_encoding)
```

```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler , MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split

df = pd.read_csv('sample_data.csv')
df_scaler = df.copy()

le= LabelEncoder()
df_scaler['Gender_encoded'] = le.fit_transform(df_scaler['Gender'])

df_scaler = pd.get_dummies(df_scaler, columns=['City'], dtype=int)

cols_to_scale = [
    'Gender_encoded',
    'City_Bangalore',
    'City_Chennai',
    'City_Delhi',
    'City_Mumbai'
]

# print(df_scaler)

Standard_Scaler = StandardScaler()
Standard_Scaled = Standard_Scaler. fit_transform(df_scaler[cols_to_scale])

print("standard scaler output:")
print(pd.DataFrame(Standard_Scaled, columns=cols_to_scale))

Minmax_Scaler = MinMaxScaler()
Minmax_Scaled = Minmax_Scaler.fit_transform(df_scaler[cols_to_scale])

print("Min Max scaler output:")
print(pd.DataFrame(Minmax_Scaled, columns=cols_to_scale))

X = df_scaler[['Gender_encoded']]
Y = df_scaler[['City_Bangalore', 'City_Chennai', 'City_Delhi', 'City_Mumbai']]

X_train , X_test , Y_train , Y_test =train_test_split(X, Y, test_size= 0.2 , random_state=42)

print("X-Train", X_train)
print("X-Test", X_test)
print("Y-Train", Y_train)
print("Y-Test", Y_test)






```
