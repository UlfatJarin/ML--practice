```python
import pandas as pd 

data = {
    "Name": ["Aban", "Apil", "Alit", "Shan", "Om", "Riya", "Khan", "Mitu"],
    "Age": [25, None, 44, 23, None, 29, 35, None],
    "Department": ["IT", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR"],
    "Salary": [50000, 60000, 70000, None, None, 65000, 72000, 58000],
    "Experience": [2, 4, 10, 1, None, 5, 8, 3]
}

df = pd.DataFrame(data)

print(df)

#check empty by number
print(df.isnull().sum())

#check empty by percentage 
print(df.isnull().mean()* 100)

#ignore unfinshed data
print(df.dropna())

#Nan is fill with mean value
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary']).mean()
df['Experience'] = df['Experience'].fillna(df['Experience'].mean())

print(df)
```
