```python
import pandas as pd

# Fixed dictionary
data = {
    'name': ['Paban', 'Kapil', 'Lalit', 'Ishan', 'Om'],
    'age': [25, None, 44, 23, None],
    'salary': [50000, 60000, 70000, None, None]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Origenal Dataframe")
print(df)

print(df.isnull().sum())

df_drop = df.dropna()
print(df_drop)

df['age'].fillna(df['age'].mean(),inplace=True)
df['salary'].fillna(df['salary'].mean(),inplace=True)

print(df)
```
