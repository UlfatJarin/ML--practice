```python
import pandas as pd

# Fixed dictionary
data = {
    'name': ['Paban', 'Kapil', 'Lalit', 'Ishan', 'Om'],
    'Age': [25, None, 44, 23, None],
    'Salary': [50000, 60000, 70000, None, None]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Origenal Dataframe")
print(df)

print(df.isnull().sum())

df_drop = df.dropna()
print(df_drop)

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

print(df)

print(df.isnull().mean() *100)
```

```python
##lavel Encoding 
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df = pd.read_csv("sample_data.csv")

df_label = df.copy()

le = LabelEncoder()
df_label['Gender_Encoded'] = le.fit_transform(df_label['Gender'])
df_label['Passed_Encoded'] = le.fit_transform(df_label['Passed'])
''' fit_transfer
    fit - learn 
    trasfer - then convert  as number like (0,1)
'''

print('\n Label Encoded Data')
print(df_label[['Name' ,'Gender', 'Gender_Encoded', 'Passed', 'Passed_Encoded']])


## One-Hot Encoding

df_encoded = pd.get_dummies(df_label, columns=['City'], dtype=int)
'''
get_dummies
convert index in small binary columes
'''
print('\nOne_Hot Encoded Data(City)')
print(df_encoded)
```

```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler , MinMaxScaler
from sklearn.model_selection import train_test_split

'''
#StandardScaler = used for standardization (mean = 0, std = 1)
#MinMaxScaler   = used for scaling values between 0 and 1 
#train_test_split â†’ used to split data into train & test sets
'''

## train_test_split == for  train machine 

data = { 
    'StudyHours': [1,2,3,4,5],
    'TestScore' :[ 40,50, 60, 70,80]
}

df= pd.DataFrame(data)
standard_scaler = StandardScaler()
standard_scaled = standard_scaler. fit_transform(df)

print("Standard Scaler OutPut :")
print(pd.DataFrame(standard_scaled, columns =['StudyHours' , 'TestScore']))

minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df)

print("\n MinMax Scaled Output:")
print(pd.DataFrame(minmax_scaled, columns=['StudyHours' , 'TestScore']))

X= df[['StudyHours']]
Y = df[['TestScore']]

X_train , X_test , Y_train , Y_test = train_test_split(X,Y, test_size=0.2 ,random_state=42)
print("Training Data")
print(X_train)

print("Test Data ")
print(X_test)

print("Training Data")
print(Y_train)

print("Test Data ")
print(Y_test)

```
