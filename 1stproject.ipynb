```python
# project1 

import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

data = pd.read_csv("student.csv")

X= data[['Hours']] 
y= data['Score'] #target column

model = LinearRegression()
model.fit(X,y)

predicted_score = model.predict(X)

mae = mean_absolute_error(y, predicted_score)
mse = mean_squared_error(y, predicted_score)
rmse = np.sqrt(mse)

print("MAE : ", mae)
print("MSE : ", mse)
print("RMSE : ",rmse)

new_hour = float(input("enter a Hour= "))
new_pred = model.predict([[new_hour]])
print(f"Prediction for {new_hour} is score ={new_pred}")
```

```python
#2nd project 
'''
3step formula

1st = X and y are numeric? = yes
    - scatter plot 
      relationship

2nd = Is one column a catagory ?
    - Bar plot /count Plot

3rd = want to see a distribution shape ?
    - histogram / KDE Plot / Box plot


1.0= perfect 
0.0 = model knows better 
<0 = model is worse then avarage
    
'''
'''
in project 

1st - find problem 

problem- 
 1 . if hour of studies will efect in ther final exam ?
    - compare with there weakly study hour
   
'''

import pandas as pd 
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt 

#load dataset 
data = pd.read_csv("student_performance_1000.csv")


#input data
X= data[['Study_Hours_per_Week']]
y= data['Final_Score']

#train model 
model = LinearRegression()

model.fit(X,y)

predicted_score = model.predict(X)

#valid regression metrics
mae = mean_absolute_error(y, predicted_score)
mse = mean_squared_error(y, predicted_score)
rmse = np.sqrt(mse)
r2 = r2_score(y, predicted_score)

#show result
print('MAE : ', round (mae, 2))
print('MSE : ', round(mse, 2))
print('RMSE : ', round(rmse, 2))
print('R2 SCore(Model Accuracy) : ', round(r2,2)) #closer to 1 = better 

#histogram
plt.figure(figsize=(10,6))
plt.hist(data["Final_Score"], bins =10 ,color='skyblue' ,edgecolor= 'black') #bins use in hish. means how many bar
plt.title("Distibution of FINAL EXAM SCORES")
plt.xlabel("Final exam score")
plt.ylabel("Number of student")
plt.grid(True)
plt.show()

#scatter - Regression line 
plt.figure(figsize=(10,6))
plt.scatter(X,y,color='blue' ,label='Actual Scores') 
plt.plot(X, predicted_score , color="red", label= "Predicted Scores (Regression Line)")
plt.title("Model prediction VS Actual Score")
plt.xlabel("Study Hours per week")
plt.ylabel("Final Output")
plt.grid(True)
plt.show()

new_hour =9
predicted_new_score = model.predict([[new_hour]])
print(f"Predicted Final Score for {new_hour} Hours is {predicted_new_score}")


```
